##10. 가우시안 RBF 커널, 시그모이드 커널을 활용한 커널 트릭 비선형 분류

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm, datasets
```

```python
X, y = datasets.make_moons(n_samples=100, noise=0.1, random_state=0)

plt.scatter(X[:,0], X[:,1], c=y, cmap=plt.cm.Paired)
plt.show()
```

```python
# 가우시안 RBF 커널
rbf_model = svm.SVC(kernel='rbf', gamma=2, C=1)
rbf_model.fit(X, y)

# 시그모이드 커널
sigmoid_model = svm.SVC(kernel='sigmoid', C=1)
sigmoid_model.fit(X, y)
```

```python
# 가우시안 RBF 커널 모델 시각화
plt.subplot(121)
plt.scatter(X[:, 0], X[:, 1], c=y, s=30, cmap=plt.cm.Paired)
xlim = plt.xlim()
ylim = plt.ylim()

xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 200),
                     np.linspace(ylim[0], ylim[1], 200))
Z = rbf_model.decision_function(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, cmap=plt.cm.Paired, alpha=0.5)
plt.scatter(rbf_model.support_vectors_[:, 0], rbf_model.support_vectors_[:, 1], s=100,linewidth=1, facecolors='none', edgecolors='k')

plt.title('Gaussian RBF Kernel')
```

```python
plt.subplot(122)
plt.scatter(X[:, 0], X[:, 1], c=y, s=30, cmap=plt.cm.Paired)
Z = sigmoid_model.decision_function(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, cmap=plt.cm.Paired, alpha=0.5)
plt.scatter(sigmoid_model.support_vectors_[:, 0], sigmoid_model.support_vectors_[:, 1], s=100,
linewidth=1, facecolors='none', edgecolors='k')

plt.title('Sigmoid Kernel')

plt.xlim(xlim)
plt.ylim(ylim)
plt.show()
```
