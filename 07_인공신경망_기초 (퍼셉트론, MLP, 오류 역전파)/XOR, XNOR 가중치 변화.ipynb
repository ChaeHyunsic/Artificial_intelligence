기본 선언문

```python
import numpy as np

W = np.array([0, 0, 0, 0])
learning_rate = 0.01

def activation(s):
    if s >= 0: return 1
    else: return 0

def out(polyX) :
    return activation (W.dot(polyX))

def train(x0, x1, target):
    global W
    polyX = np.array([x0, x1, x0*x1, 1])                                          # 다항변환식 적용
    y = out(polyX)

    ### 예측이 맞으면 아무것도 하지 않음-------------------------------------
    if target == y: return 0               # 가중치가 변경되지 않았음을 반환
    ### 예측이 틀리면 학습 실시---------------------------------------------
    W = W - learning_rate * polyX * (y - target)    # 오차에 비례하여 변경
    print('X : [{} {}], target : {}, y : {}, error:{}, W:{}'.format(x0, x1, target, y, (y - target), W))

    return 1                               # 가중치가 변경되었음을 반환

def predict(inputs):
    outputs = []
    for x in inputs:
        polyX = np.array([x[0], x[1], x[0]*x[1], 1])
        outputs.append (out(polyX))

    return outputs
```

XOR

```python
adjusted = 0
for i in range(100):
    adjusted += train(0, 0, 0)    # 훈련 데이터 1
    adjusted += train(0, 1, 1)    # 훈련 데이터 2
    adjusted += train(1, 0, 1)    # 훈련 데이터 3
    adjusted += train(1, 1, 0)    # 훈련 데이터 4
    print("iteration -------------", i)
    if not adjusted: break  # 모든 훈련에 대해 가중치 변화 없으면 학습종료
    adjusted = 0

X = [[0, 0], [0, 1], [1, 0], [1,1]]
yhat = predict(X)

print('')
print('result ----------------')
print('x0 x1  y')
for i in range(len(X)):
    print('{0:2d} {1:2d} {2:2d}'.format(X[i][0], X[i][1], yhat[i]))
```
